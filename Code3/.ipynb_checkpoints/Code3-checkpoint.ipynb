# Libraries

```python
# Import Allowed Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
```

# Load and Overview

```python
data = pd.read_csv("iris.data", header=None, names=["Sepal Length", "Sepal Width", "Petal Length", "Petal Width", "Class"])
```

# Train and Test Split

```python
if False:
    # Shuffle data one time and save it for ever!
    data = data.sample(frac=1).reset_index()
    
    # 80% train, 20% test
    train_data = data.head(int(len(data)*0.8))
    test_data  = data.tail(int(len(data)) - int(len(data)*0.8))
    
    # Seperate train data to 10 folds
    for i, item in train_data.iterrows():
        train_data.at[i, "Fold"] = int(i/(int(len(data)/10))) 
    
    train_data.to_csv("train_data.csv")
    test_data.to_csv("test_data.csv")
else:
    # Load shuffled data
    train_data = pd.read_csv("train_data.csv").drop(columns = ["Unnamed: 0"])
    test_data = pd.read_csv("test_data.csv").drop(columns = ["Unnamed: 0"])
```

# Normalization

```python
def normalize(data):
    """ Apply normalization on data"""
    data["Sepal Length"]=(data["Sepal Length"]-data["Sepal Length"].mean())/data["Sepal Length"].std()
    data["Sepal Width"]=(data["Sepal Width"]-data["Sepal Width"].mean())/data["Sepal Width"].std()
    data["Petal Length"]=(data["Petal Length"]-data["Petal Length"].mean())/data["Petal Length"].std()
    data["Petal Width"]=(data["Petal Width"]-data["Petal Width"].mean())/data["Petal Width"].std()
```

```python
normalize(train_data)
normalize(test_data)
```

# Cross Validation

```python
def distance(item1, item2):
    """ Distance of two item in the dataset """
    sum = 0
    for feature in ["Sepal Length", "Sepal Width", "Petal Length", "Petal Width"]:
        sum += (item1[feature] - item2[feature]) ** 2
    return sum ** 0.5
```

```python
def knn(k, train_part, test_item):
    """
    Use knn for predict class
    
    :param k: k in knn
    :param train_part: train dataset
    :param test_item: the item that we will predict its class
    :return: the class of test_item
    """
    distance_list = []
    for index, train_item in train_part.iterrows():
        distance_list.append({"Class": train_item["Class"], "Distance": distance(test_item, train_item)})

    nearest_list = sorted(distance_list, key=lambda x:x["Distance"]) 
    nearest_list = [nearest["Class"] for nearest in nearest_list[:k]]
    predict_class = max(set(nearest_list), key=nearest_list.count)

    return predict_class
```

```python
def knn_cross_validation(k):
    """
    Run KNN with cross validation metric
    
    :param k: k in knn
    :return: accuracy of cross validation
    """
    total_number = 0
    correct_number = 0

    # Iteration for k fold cross validation
    for validation_fold in range(10):
        validation_part = train_data[train_data["Fold"] == validation_fold]
        train_part = train_data[train_data["Fold"] != validation_fold]

        # Iteration for every data in validation set
        for index, item in validation_part.iterrows():
            
            predict_class = knn(k, train_part, item)
            real_class = item["Class"]
            
            # Evaluation
            total_number += 1
            if real_class == predict_class: correct_number += 1 

    accuracy = correct_number/total_number
    return accuracy
```

```python
k_list = []
accuracy_list = []

best_k = -1
maxim_accuracy = 0

for k in range(1, 35, 2):
    accuracy = knn_cross_validation(k)
    if accuracy >= maxim_accuracy:
        best_k = k
        maxim_accuracy = accuracy
    
    k_list.append(k)
    accuracy_list.append(accuracy)

print(f"Best k = {best_k} | Cross Validation Accuracy = {maxim_accuracy}")    
    
plt.scatter(k_list, accuracy_list)
plt.plot(k_list, accuracy_list)
plt.title("Cross validation accuracy for different k in knn")
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.show()
```

# Evaluation

```python
total_number = 0
correct_number = 0

real_label_list = []
predict_label_list = []
for index, test_item in test_data.iterrows():
    predict_class = knn(best_k, train_data, test_item)
    real_class = test_item["Class"]
    
    predict_label_list.append(predict_class)
    real_label_list.append(real_class)
    
    # Evaluation
    total_number += 1
    if real_class == predict_class: correct_number += 1 
        
print(f"Accuracy = {correct_number/total_number}.2")
print(confusion_matrix(real_label_list, predict_label_list, labels= ["Iris-setosa", "Iris-versicolor", "Iris-virginica"]))
```
